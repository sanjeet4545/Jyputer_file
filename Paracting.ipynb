```python
# Let's check how to data has been distributed each column
plt.figure(figsize=(15,10))
plotnumber=1

for col in df:
    if plotnumber<=5:
        ax = plt.subplot(3,3,plotnumber)
        sns.distplot(df[col])
        plt.xlabel(col,fontsize=12)
    plotnumber+=1
plt.tight_layout()
```

```python
# Let's check the outliers by using boxplot
plt.figure(figsize=(15,10))
plotnumber=1

for col in df:
    if plotnumber<=5:
        ax = plt.subplot(3,3,plotnumber)
        sns.boxplot(df[col])
        plt.xlabel(col,fontsize=12)
    plotnumber+=1
plt.tight_layout()
```

```python
from scipy.stats import zscore
out_features = df[['Open','High','Low','Volume']]
z = np.abs(zscore(out_features))
z
```

```python
# Thresholds
np.where(z>3)
```

```python
z.iloc[723,3]
```

```python
df1 = df[(z<3).all(axis=1)]

df1.shape
```

```python
# Shape of old and new dataframe
print('Old DataFrame',df.shape[0])
print('New DataFrame',df1.shape[0])
```

```python
print('Data Loss Percentage :',((df.shape[0] - df1.shape[0])/df.shape[0])*100)
```
